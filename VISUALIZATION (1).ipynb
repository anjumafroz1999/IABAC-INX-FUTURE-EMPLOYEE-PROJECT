### 2.UNIVARIATE ANALYSIS
Examines one variable at a time.

Helps understand distribution, central tendency, spread, and outliers.

PLOTS USED : COUNTPLOT,HISTOGRAM

COUNT PLOT : : A count plot is a type of bar chart that shows the frequency of categorical values.

HISTPLOT : A histogram is a plot that shows the distribution of a continuous numeric variable.

```python
# UNVIVARIATE ANALYSIS OF DEMOGRAPHIC FEATURES
cols = ['Age','Gender','MaritalStatus']

for col in cols:
    plt.figure(figsize=(6,4))
    
    if data[col].dtype == 'object' or data[col].dtype.name == 'category':
        # Count plot for categorical features
        sns.countplot(x=col, data=data, palette="pastel")
    else:
        # Histogram for numeric features
        sns.histplot(x=data[col], kde=True, bins=30, color="skyblue", edgecolor="black")
        
        # Add mean line
        mean_val = data[col].mean()
        plt.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
        plt.legend()
    
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

```

####  Age Distribution:

* The average employee age is ~37 years, with most clustered between 30–40 years.

* This suggests a mid‑career workforce, likely balancing experience with growth potential.

* Younger and older employees are fewer, which may influence overall performance dynamics and promotion opportunities.

#### Gender Distribution:

* The dataset shows more male employees (700) compared to female employees (500).

* This imbalance could affect diversity initiatives and may reflect industry‑specific hiring trends.

* Understanding whether performance ratings differ by gender can highlight potential bias or equity issues.

#### Marital Status Distribution:

* Married employees (550) form the largest group, followed by Single (390) and Divorced (260).

* A higher proportion of married employees may indicate workforce stability, but singles and divorced groups could show different engagement or attrition patterns.

* Linking marital status with performance ratings can reveal whether personal circumstances influence workplace outcomes.

# 4.MULTIVARIATE ANALYSIS
studying how multiple variables interact together to explain outcomes or reveal patterns.

Multiple variables → Often includes one dependent (target) and several independent (predictors).

Interrelationships → Examines how variables interact together, not just pairwise.



### 1.RELATIONSHIP BETWEEN PREDICTORS AND PERFORMANCE RATING

PLOT USED : HEAT MAP

HEAT MAP :A heat map is a data visualization that uses color intensity or gradients to represent values in a matrix or table.

```python
sns.heatmap(data[['Age','YearsWithCurrManager','EmpJobSatisfaction',
                  'EmpWorkLifeBalance','PerformanceRating']].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap: Predictors vs PerformanceRating")
plt.show()
```

OBSERVATION :
Performance ratings are largely independent of demographics (Age) and satisfaction scores.

Managerial tenure shows a small negative link, hinting at possible complacency or lack of fresh opportunities.

Work‑life balance is the only factor with a positive signal, suggesting HR policies that improve balance could indirectly support performance.

### 2. RELATIONSHIP BETWEEN EMPWORK LIFE BALANCE, AND PERFORMANCE RATING WITH RESPECT TO OVER TIME
plot used : box plot

```python
sns.boxplot(x="EmpWorkLifeBalance", y="PerformanceRating",
            hue="OverTime", data=data, palette="Set2")
```

OBSERVATIONS:

Overtime may reduce effectiveness → Continuous overtime could be counterproductive, lowering performance ratings despite extra hours.

### 3.RELATIONSHIP BETWEEN AGE, EXPERIENCE IN CURRENT ROLE AND PERFORMANCE RATING
PLOTS USED : SCATTER PLOT
SCATTERPLOT :A scatter plot is a type of graph used to display the relationship between two continuous variables.

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="Age", 
                y="ExperienceYearsInCurrentRole", 
                hue="PerformanceRating", 
                data=data, 
                palette="viridis", 
                alpha=0.7)

plt.title("Age vs Experience in Current Role (Performance Rating as Hue)")
plt.xlabel("Age")
plt.ylabel("Experience in Current Role (Years)")
plt.legend(title="Performance Rating")
plt.show()
```

INSIGHT:
Older employees tend to accumulate more years in their current role, but performance ratings are spread across all age–experience combinations, showing no strong rating bias by age or tenure.

### CHECKING SKEWNESS OF DATA
SKEWNESS:Skewness is a statistical measure that describes the asymmetry of a distribution around its mean.
If a distribution is perfectly symmetric (like a normal distribution), skewness = 0.

Positive skew (right‑skewed): The tail on the right side is longer or fatter; most values are concentrated on the left, but a few large values pull the mean upward.

Negative skew (left‑skewed): The tail on the left side is longer or fatter; most values are concentrated on the right, but a few small values pull the mean downward.

```python
for col in data.columns:
    # Only calculate skewness for numeric columns
    if pd.api.types.is_numeric_dtype(data[col]):
        skew_val = data[col].skew()
        print(f"Skewness of {col}: {skew_val}")
```

### CHECKING KURTOSIS
KURTOSIS : It is measure of peakness in data relative to a normal distribution.

Normal distribution kurtosis = 3 (often reported as “excess kurtosis = 0”).

High kurtosis (>3) → Heavy tails, more outliers, sharper peak.

Low kurtosis (<3) → Light tails, fewer outliers, flatter peak.

```python
for col in data.columns:
    if pd.api.types.is_numeric_dtype(data[col]):
        kurt_val=data[col].kurt()
        print(f"Kurtosis of {col} : {kurt_val}")
```

ExperienceYearsAtThisCompany → 4.06  
Very high kurtosis, meaning the distribution has heavy tails — most employees cluster around a few years, but some stay much longer, creating extreme values.

YearsSinceLastPromotion → 3.54  
Also highly kurtotic, indicating that while many employees have short gaps between promotions, a few have very long delays, producing strong outliers.

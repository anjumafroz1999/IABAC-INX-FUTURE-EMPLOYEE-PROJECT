#### 1. Encoding and Column Reduction  
Categorical variables were encoded using label,manual and frequency encoding techniques, converting all categorical columns into numeric form. The unique identifier EmployeeNumber was dropped, reducing the dataset to 27 features.

#### 2. Outlier Detection and Imputation  
Box plots revealed outliers in tenure‑related variables (total experience, training times last year, years at company, years in current role, years with current manager, years since last promotion). Outliers were handled using the Interquartile Range (IQR) method, with median imputation to avoid bias from right‑skewed distributions.

#### 3. Feature Transformation  
Skewness correction was applied: Years Since Last Promotion underwent log transformation. Continuous variables (age, distance from home, hourly rate, salary hike %, total experience, work experience, training times, tenure features, and log‑transformed promotion variable) were standardized using StandardScaler.

# FEATURE SELECTION
#### 1. Correlation and Duplicate Checks
Correlation analysis via heatmap confirmed no highly correlated features, so no variables were dropped. Duplicate checks returned clean results

#### 2. Dimensionality Reduction (PCA)
Principal Component Analysis was performed to reduce dimensionality. The scree plot indicated that 18 components retained ~95% of variance. These components were used to create a reduced dataset. The PCA output (NumPy array) was converted into a Pandas DataFrame and saved as a CSV file for subsequent model building.

#### 6. Model Evaluation and Selection
